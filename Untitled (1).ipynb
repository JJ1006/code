```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the file
data_path = "data_1.txt"
data = np.loadtxt(data_path)

# Extract x and y values from the data
x = data[0, :]  # First row represents x-values
y_true = data[1, :]  # Second row represents y-values

# Define the 4th-degree polynomial function
def polynomial(x, q, r, s, m, b):
    """
    Computes the 4th-degree polynomial function.
    
    Parameters:
    x (array): Input values
    q, r, s, m, b (float): Coefficients of the polynomial
    
    Returns:
    array: Computed polynomial values
    """
    return q * x**4 + r * x**3 + s * x**2 + m * x + b

# Compute the gradients for gradient descent
def compute_gradients(x, y_true, params):
    """
    Computes the gradients of the loss function with respect to each parameter.
    
    Parameters:
    x (array): Input values
    y_true (array): Actual y values
    params (array): Current parameter estimates
    
    Returns:
    array: Computed gradients for each parameter
    """
    q, r, s, m, b = params
    n = len(x)
    y_pred = polynomial(x, q, r, s, m, b)  # Predicted y values
    error = y_true - y_pred  # Compute residual error
    
    # Compute gradients for each parameter using Mean Squared Error (MSE) derivative
    dq = -2 * np.sum(error * x**4) / n
    dr = -2 * np.sum(error * x**3) / n
    ds = -2 * np.sum(error * x**2) / n
    dm = -2 * np.sum(error * x) / n
    db = -2 * np.sum(error) / n
    
    return np.array([dq, dr, ds, dm, db])

# Gradient Descent Optimization
def gradient_descent(x, y_true, alpha=0.01, num_iterations=10000, tolerance=1e-6):
    """
    Performs gradient descent optimization to find the best-fit polynomial coefficients.
    
    Parameters:
    x (array): Input values
    y_true (array): Actual y values
    alpha (float): Learning rate (default=0.01)
    num_iterations (int): Maximum number of iterations (default=10000)
    tolerance (float): Minimum improvement threshold to stop early (default=1e-6)
    
    Returns:
    tuple: Optimized parameters and loss history
    """
    params = np.zeros(5)  # Initialize parameters with zeros
    loss_history = []  # Store loss values for tracking progress
    
    for i in range(num_iterations):
        gradients = compute_gradients(x, y_true, params)  # Compute gradients
        params -= alpha * gradients  # Update parameters using gradient descent
        
        # Compute loss (Mean Squared Error - MSE)
        loss = np.sum((y_true - polynomial(x, *params))**2)
        loss_history.append(loss)
        
        # Early stopping condition: Stop if the loss improvement is very small
        if i > 0 and abs(loss_history[-2] - loss_history[-1]) < tolerance:
            break  
    
    return params, loss_history

# Train the model using gradient descent
params_opt, loss_history = gradient_descent(x, y_true)
q_opt, r_opt, s_opt, m_opt, b_opt = params_opt  # Extract optimized parameters

# Generate predictions using optimized parameters
y_pred_opt = polynomial(x, q_opt, r_opt, s_opt, m_opt, b_opt)

# Plot actual data vs optimized polynomial fit
plt.scatter(x, y_true, label="Actual Data", alpha=0.5)  # Scatter plot for actual data
plt.plot(x, y_pred_opt, color='red', label="Optimized Polynomial Fit")  # Line plot for model
plt.legend()
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("4th Degree Polynomial Fit with Gradient Descent")
plt.grid(True)
plt.show()

# Print the final optimized polynomial equation
print(f"Optimized Polynomial: y = {q_opt:.4f}x^4 + {r_opt:.4f}x^3 + {s_opt:.4f}x^2 + {m_opt:.4f}x + {b_opt:.4f}")
```
